```python
from tensorflow.keras import datasets
# CIFAR-10データを読み込む
(x_train, t_train), (x_test, t_test) = datasets.cifar10.load_data()
# データのピクセル値を0～1の範囲に変換
x_train = x_train/ 255.0

# 使用する画像の枚数
batch_size = 16
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def draw(x):
    '''
    描画を行う関数
    
    '''
    plt.figure(figsize=(8, 8)) # 描画エリアは8×8インチ
    pos = 1                    # 画像の描画位置を保持
    # 画像の枚数だけ描画処理を繰り返す
    for i in range(len(x)):
        plt.subplot(4, 4, pos) # 4×4の描画領域のpos番目の位置
        plt.imshow(x[i])       # インデックスiの画像を描画
        plt.axis('off')        # 軸目盛は非表示
        pos += 1
    plt.show()
```

```python
# オリジナルの画像を表示
draw(x_train[0:batch_size])
```

```python
# ImageDataGeneratorのインポート
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 回転処理　最大90度
datagen = ImageDataGenerator(rotation_range=90)
g = datagen.flow(           # バッチサイズの数だけ拡張データを作成
    x_train, t_train, batch_size, shuffle=False)
x_batch, t_batch = g.next() # 拡張データをリストに格納
draw(x_batch)               # 描画
```

```python
# 平行移動　最大0.5
datagen = ImageDataGenerator(width_shift_range=0.5)
g = datagen.flow(           # 拡張データを作成
    x_train, t_train, batch_size, shuffle=False)
x_batch, t_batch = g.next() # 拡張データをリストに格納
draw(x_batch)               # 描画
```

```python
# 垂直移動　最大0.5
datagen = ImageDataGenerator(height_shift_range=0.5)
g = datagen.flow(           # 拡張データを作成
    x_train, t_train, batch_size, shuffle=False)
x_batch, t_batch = g.next() # 拡張データをリストに格納
draw(x_batch)               # 描画
```

```python
# ランダムに拡大　最大0.5
datagen = ImageDataGenerator(zoom_range=0.8)
g = datagen.flow(           # 拡張データを作成
    x_train, t_train, batch_size, shuffle=False)
x_batch, t_batch = g.next() # 拡張データをリストに格納
draw(x_batch)               # 描画
```

```python
# 左右をランダムに反転
datagen = ImageDataGenerator(horizontal_flip=True)
g = datagen.flow(           # 拡張データを作成
    x_train, t_train, batch_size, shuffle=False)
x_batch, t_batch = g.next() # 拡張データをリストに格納
draw(x_batch)               # 描画
```

```python
# 上下をランダムに反転
datagen = ImageDataGenerator(vertical_flip=True)
g = datagen.flow(           # 拡張データを作成
    x_train, t_train, batch_size, shuffle=False)
x_batch, t_batch = g.next() # 拡張データをリストに格納
draw(x_batch)               # 描画
```

```python
# 画像のチャンネルをランダムに移動　最大0.7
datagen = ImageDataGenerator(channel_shift_range=0.7)
g = datagen.flow(           # 拡張データを作成
    x_train, t_train, batch_size, shuffle=False)
x_batch, t_batch = g.next() # 拡張データをリストに格納
draw(x_batch)               # 描画
```
