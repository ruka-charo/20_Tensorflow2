```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import datasets
%matplotlib inline

# CIFAR-10
(x_train, t_train), (x_test, t_test) = datasets.cifar10.load_data()
# 
print('x_train:', x_train.shape, 't_train:', t_train.shape)
print('x_test :', x_test.shape, 't_test :', t_test.shape)

# 
num_classes = 10 # 
pos = 1          # 

# 
for target_class in range(num_classes):
    # 
    target_idx = []
    
    # i
    for i in range(len(t_train)):
        # i0target_class
        if t_train[i][0] == target_class:
            # itargetIdx
            target_idx.append(i)
    
    np.random.shuffle(target_idx) # i
    plt.figure(figsize=(20, 20))  # 253
    
    # 10
    for idx in target_idx[:10]:   
        plt.subplot(10, 10, pos)  # 1010pos
        plt.imshow(x_train[idx])  # Matplotlibimshow()
        pos += 1

plt.show()
```
