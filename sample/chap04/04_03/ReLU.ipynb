```python
# ReLU関数の出力をグラフにする
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# ReLU関数
def relu(x):
    y = np.maximum(0, x)
    return y

# -6から6までの値0.1刻みで等差数列を生成
x = np.arange(-6, 6, 0.1)
# xにReLU関数を適用
y = relu(x)

plt.plot(x, y)
plt.xlim(-1, 1)
plt.ylim(-0.1, 1)
plt.grid()
plt.show()
```
